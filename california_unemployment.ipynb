### **Note on Data Observations:**

**Presence of extreme (0 and 100%) Values:** The columns for employment, unemployment, labor force, and unemployment rate contain some entries with 0 values. The unemployment rate ranges from a minimum of 0% to a maximum of 100%. These zero values and 100% indicate ambiguous or missing data. Such extremes are plausible within sub-county zones, potentially reflecting actual conditions. However, outside of these specific areas, they suggest a high likelihood of data inaccuracies or omissions.

```python
# Filter rows where 'Unemployment Rate' equals 0
unemployment_zero_data = data[data['Unemployment Rate'] == 0]

print(unemployment_zero_data.shape)
# Find unique categories in the 'Area Type' column of the filtered DataFrame
unique_area_types = unemployment_zero_data['Area Type'].unique()

# Print the unique categories
print(unique_area_types)
```

We found 0% unemployment rate in Sub-County Place cateogory only.

Finding category of 'Area Type' column where the unemployment rate is 100%.

```python
# Filter rows where 'Unemployment Rate' equals 0
unemployment_hundred_data = data[data['Unemployment Rate'] == 100]

print(unemployment_hundred_data.shape)
# Find unique categories in the 'Area Type' column of the filtered DataFrame
unique_area_types = unemployment_hundred_data['Area Type'].unique()

# Print the unique categories
print(unique_area_types)
```

We found the 100% unemployment percentage is within the Sub County Zone.

Checking the number of rows where the number of employment is zero and category of the 'Area Type' column:

```python
zero_employment = data[data['Employment'] == 0]
print(zero_employment.shape)

print(zero_employment['Area Type'].unique())
```

We found the employment column has 0 values in 'Sub-County Place' column only.

Checking the category of Area Type column where Labor force value is 0:

```python
zero_labor_force = data[data['Labor Force'] == 0]
print(zero_labor_force.shape)

print(zero_labor_force['Area Type'].unique())
```

### **Consideration and Decision**

It has been observed that entries in the Labor Force, Employment, Unemployment, Unemployment Rate column are marked as 0 exclusively for areas classified under the 'Sub-County Place' category. In addition, maximum value of Unemployment rate is 100%.

While initially suggestive of missing data, the unique characteristics of these areas—their notably smaller scale compared to other categories—introduce a plausible rationale for such figures.

Given the ambiguity of these data points and the potential for misinterpretation without comprehensive research, a decision has been made to exclude entries from the 'Sub-County Place' category in our analysis. This approach aims to maintain the integrity and reliability of our project's findings, acknowledging the limitations of the available data.

### **Data Cleaning**

Since 'Status', 'Seasonally Adjusted (Y N)', and 'Period' column have only 1 unique value, we do not have the importance of them in our EDA and. visualization task. Therefore, dropping the three columns.

```python
CA_unemployment = data.drop(['Status', 'Seasonally Adjusted (Y N)', 'Period'], axis = 1)
```
